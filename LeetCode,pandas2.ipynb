```python
data = [[1, 'Alice'], [2, 'Bob'], [3, 'Alex'], [4, 'Donald'], [7, 'Lee'], [13, 'Jonathan'], [19, 'Elvis']]
users = pd.DataFrame(data, columns=['id', 'name']).astype({'id':'Int64', 'name':'object'})
data = [[1, 1, 120], [2, 2, 317], [3, 3, 222], [4, 7, 100], [5, 13, 312], [6, 19, 50], [7, 7, 120], [8, 19, 400], [9, 7, 230]]
rides = pd.DataFrame(data, columns=['id', 'user_id', 'distance']).astype({'id':'Int64', 'user_id':'Int64', 'distance':'Int64'})
```

```python
# 가장 직관적인 풀이
user_id = users['id'].unique()
dic = dict()
name_lst = []
distance_lst = []
for i, id in enumerate(user_id):
  name_lst.append(users[users['id']==id]['name'].values[0])
  distance_lst.append(sum(new_rides[new_rides['user_id']==id]['distance']))
dic['name'] = name_lst
dic['distance'] = distance_lst
dic
```

```python
df = pd.DataFrame(dic)
df.sort_values(axis=0, by=['distance','name'], ascending=[False, True])
```

```python
# groupby와 merge를 이용하면
df = rides.groupby(['user_id'])['distance'].sum().reset_index(name='travelled_distance')
df
```

```python
ddf = df.merge(users, how='right', left_on='user_id', right_on='id')
ddf.sort_values( by=['travelled_distance','name'], ascending=[False,True], inplace=True )
ddf = ddf.fillna(0)
ddf
```
